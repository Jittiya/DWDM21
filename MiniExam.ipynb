## ขั้นตอนแรก ทำการรวมตาราง

### หาว่า การทำ dropna() ทำให้ข้อมูลหายไปกี่ %

````python
removed = merged_start.shape[0] - merged_start.dropna().shape[0]

print(f'size before drop = {merged_start.shape[0]}')
print(f'size after drop = {merged_start.dropna().shape[0]}')
print(f'we lost {100*(removed/merged_start.shape[0])}% of data')
````

จากการ dropna() เพื่อลบค่า missing value หรือค่า NaN 

พบว่าทำให้ข้อมูลหายไป 0.010915766177628008% ซึ่งอยู่ในเกณฑ์ที่ยอมรับได้ 

จึงเลือกใช้วิธีนี้ในการลบค่า missing value

````python
merged_start_remove = merged_start.dropna()     # ตั้งชื่อตารางใหม่
merged_start_remove
````

note รหัสสนามบินต้นทาง =  ขาออก\

# ให้นักศึกษาคิดโจทย์เองอีก 1 ข้อ

(จากตาราง merged_start_remove) สายการบินไหนที่มีค่าเฉลี่ยของ ความล่าช้าขาออก สูงที่สุด

````python
import numpy as np
````

````python
np.mean([10,100,70,150])
````

````python
merged_mean = merged_start_remove[merged_start_remove['ความล่าช้าขาออก']>=np.mean([10,100,70,150])]
merged_mean
````

````python
merged_mean.groupby('สายการบิน').mean()
````

````python
Exam_free = merged_mean.groupby('สายการบิน').mean().sort_values(by=['ความล่าช้าขาออก'],ascending=False)
````

````python
Exam_free.iloc[:1,:]
````
